```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#image is a numpy array
im = cv2.imread("stinkbug.png")
plt.imshow(im)
sourcePoints = np.random.uniform((0,0), (im.shape[1], im.shape[0]), (10, 2))
targetPoints = np.random.uniform((0,0), (im.shape[1], im.shape[0]), (10, 2))
plt.scatter( sourcePoints[:,0], sourcePoints[:,1] )
plt.scatter( targetPoints[:,0], targetPoints[:,1] )
```
